### Antibody-drug conjugates or ADCs are a class of biopharmaceutical drugs designed as a targeted therapy for treating cancer. Unlike chemotherapy, ADCs are intended to target and kill tumor cells while sparing healthy cells. 

### These are proteins produced in the lab. So the generative AI should generate proteins that will bind effectively with specific targets such as CD33

```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
# traget search for proteins
target = new_client.target
target_query = target.search('Smoothened homolog')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
sele = targets.target_chembl_id[0]
sele
```

```python
act = new_client.activity
res = act.filter(target_chembl_id=sele).filter(standard_type='IC50')
```

```python
df = pd.DataFrame.from_dict(res)
df
```

```python
df.to_csv('Smoothened_homolog_bioactivity_data.csv', index=False)
```
