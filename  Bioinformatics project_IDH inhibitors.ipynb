```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
# traget search for proteins
target = new_client.target
target_query = target.search('IDH')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
selected_target = targets.target_chembl_id[1]
selected_target
```

```python
activity = new_client.activity
result = activity.filter(target_chembl_id=selected_target).filter(standard_type='IC50')
```

```python
actv = new_client.activity
res = actv.filter(target_chembl_id='CHEMBL2007625').filter(standard_type='IC50')
```

```python
df = pd.DataFrame.from_dict(result)
df
```

```python
df2 = pd.DataFrame.from_dict(res)
df2
```

```python
df.to_csv('Isocitrate_dehydrogenase[NADP]_mitochondrial_IDH2_bioactivity_data.csv', index=False)
df2.to_csv('Isocitrate_dehydrogenase[NADP]_cytoplasmic_IDH1_bioactivity_data.csv', index=False)
```
