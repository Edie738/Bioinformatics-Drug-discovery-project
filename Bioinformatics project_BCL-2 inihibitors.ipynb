```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
target = new_client.target
target_query = target.search('BCL-2')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
indexs = targets[targets.organism!='Homo sapiens'].index
indexs
```

```python
indexs2 = targets[targets.score < 10].index
indexs2
```

```python
targets.drop(indexs2, inplace=True)
targets
```

```python
sele = targets.target_chembl_id[24]
sele
```

```python
act = new_client.activity
res = act.filter(target_chembl_id=sele).filter(standard_type='IC50')
```

```python
df = pd.DataFrame.from_dict(res)
df
```

```python
df.to_csv('Induced myeloid leukemia cell differentiation protein Mcl-1_BCL2L3_bioactivity_data.csv', index=False)
```
